```python
import numpy as np
import os
from sentence_transformers import SentenceTransformer
from datasets import load_dataset
import pandas as pd

# Initialize the model if it's not already defined
# This ensures `model` is available within this cell's scope.
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# --- Initialize and Prepare DataFrame for Embedding Generation ---
# Load the dataset
dataset = load_dataset("CShorten/ML-ArXiv-Papers")

# Convert to pandas DataFrame and select relevant columns
df = dataset["train"].to_pandas()
df = df[['title', 'abstract']]
df = df.head(50000) # Limiting to 50,000 rows as per previous steps

# Combine title and abstract into a single 'paper_text' column
df["paper_text"] = df["title"] + " " + df["abstract"]

# Clean the 'paper_text' column by removing newlines and stripping whitespace
df["paper_text"] = df["paper_text"].str.replace("\\n", " ", regex=False)
df["paper_text"] = df["paper_text"].str.strip()

print("DataFrame 'df' initialized and prepared within this cell.")
# --- End df initialization code ---


index_path = "./data/arxiv_embeddings.npy" # Changed path to be relative to current directory

if os.path.exists(index_path):
    print("Loading saved embeddings")
    embeddings = np.load(index_path)
else:
    print("Generating embeddings")
    # Ensure the directory exists before saving
    os.makedirs(os.path.dirname(index_path), exist_ok=True)
    embeddings = model.encode(
        df["paper_text"].tolist(),
        batch_size=32,
        show_progress_bar=True
    )
    np.save(index_path, embeddings)
    print("Embeddings saved successfully!")

# The following line was incorrect and redundant, as embeddings are saved
# within the else block and 'embedding' (singular) refers to a single sample.
# np.save(index_path, embedding)

print("Embeddings successfully saved to disk!")
```

### Named Entity Recognition (NER)
Let's add Named Entity Recognition to identify key entities like persons, organizations, and locations within the paper abstracts. We'll use a pre-trained NER model from the `transformers` library.

```python
from transformers import pipeline

# Initialize the NER pipeline with a pre-trained model
# 'dslim/bert-base-NER' is a common and effective choice for general NER
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", tokenizer="dslim/bert-base-NER", aggregation_strategy="simple")

print("NER pipeline initialized!")
```

### Demonstrate NER on a sample text
Let's quickly test the NER pipeline on a sample abstract.

```python
sample_abstract_for_ner = df.iloc[I[0][0]]['abstract']
ner_results = ner_pipeline(sample_abstract_for_ner)

print("Sample Abstract:")
print(sample_abstract_for_ner[:500] + "...\n")
print("Detected Entities:")
for entity in ner_results:
    print(f"  - {entity['word']} ({entity['entity_group']})")
```

### Integrate NER into `search_and_summarize`
Now, let's update the `search_and_summarize` function to also display the detected named entities for each paper.

```python
def search_and_summarize_with_ner(query, k=5):
    D, I = search_paper(query, k)
    for score, idx in zip(D[0], I[0]):
        print("Similarity score:", score)
        print("Title:", df.iloc[idx]["title"])
        print("Abstract (Preview):" + df.iloc[idx]["abstract"][:500] + "...\n")

        # Extract keywords for the abstract
        paper_abstract = df.iloc[idx]["abstract"]
        extracted_keywords = kw_model.extract_keywords(paper_abstract, keyphrase_ngram_range=(1, 3), stop_words=None, top_n=5)
        print("Keywords:", [keyword[0] for keyword in extracted_keywords])
        print()

        # Perform Named Entity Recognition
        ner_results = ner_pipeline(paper_abstract)
        print("Detected Entities:")
        if ner_results:
            for entity in ner_results:
                print(f"  - {entity['word']} ({entity['entity_group']})")
        else:
            print("  No entities detected.")
        print()

        # Generate the summary using the AI pipeline
        summary = summarizer(paper_abstract, max_length=120, min_length=40)

        print("AI SUMMARY:")
        print(summary[0]["generated_text"])
        print("-" * 80) # Adds a nice dividing line between papers

# Call the updated function
search_and_summarize_with_ner("Deep learning in medical science",k=5)
```

### Extracting diverse keywords
To get a more diverse set of keyphrases, you can use the Maximum Marginal Relevance (MMR) algorithm by setting `use_mmr=True` and adjusting the `diversity` parameter. A higher `diversity` value means more diverse keyphrases, while a lower value prioritizes similarity to the document.

```python
keywords_diverse = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 3), stop_words=None, use_mmr=True, diversity=0.7)
print("Diverse Keyphrases (diversity=0.7):")
print(keywords_diverse)

keywords_less_diverse = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 3), stop_words=None, use_mmr=True, diversity=0.2)
print("Less Diverse Keyphrases (diversity=0.2):")
print(keywords_less_diverse)
```
